# Tweets misógino

```python
txt = '/content/drive/MyDrive/Mini-proyecto Texto/misocorpus-misogyny.txt'
with open(txt) as archivo:
  lines = archivo.readlines()
```

```python
IDs = []
for i in lines:
  IDs.append(i.strip())
  #print(i.strip())
```

```python
IDs[:5]
```

```python
#id -
#full_text -
#created_at -
#geo -
#coordinates -
#user.followers_count -
#user.friends_count -
#user.favourites_count -
#user.verified -
#user.description -
#lang -
#tweets_IDs = [1206924075374956547, 1210912199402819584]

Inicio = time()

tweets_df2 = pd.DataFrame()
for i in IDs:
  try:
    info_tweet = api.get_status(i, tweet_mode="extended")  
  except:
    pass
  tweets_df2 = tweets_df2.append(pd.DataFrame({'ID': info_tweet.id,
                                             'Tweet': info_tweet.full_text,
                                             'Creado_tweet': info_tweet.created_at,
                                             'Geo_usuario': info_tweet.geo,
                                             'Coor_usuario': info_tweet.coordinates,
                                             'Seguidores_usuario': info_tweet.user.followers_count,
                                             'Amigos_usuario': info_tweet.user.friends_count,
                                             'Favoritos_usuario': info_tweet.user.favourites_count,
                                             'Descripcion_usuario': info_tweet.user.description,
                                             'Verificado_usuario': info_tweet.user.verified,
                                             'Idioma': info_tweet.lang}, index=[0]))
  tweets_df2 = tweets_df2.reset_index(drop=True)

Final = time()
Tiempo_total = Final - Inicio

print("\nTiempo de entrenamiento: {:.0f}h {:.0f}m {:.0f}s.".format(Tiempo_total // (60 * 60),
                                                                  Tiempo_total // 60,
                                                                  Tiempo_total % 60))

```

```python
misogino['Tipo'] = np.where(len(tweets_df2)<-100, 'misogino', 'misogino')
```
